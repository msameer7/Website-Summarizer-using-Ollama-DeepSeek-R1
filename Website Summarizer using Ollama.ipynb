```python
from scraper import fetch_website_contents
```

```python
system_prompt = """
You are a witty assistant that summarizes websites.

Provide a concise summary of the main content.
Ignore navigation menus, ads, and page furniture.

You may add light humor, but do not invent facts that are not present on the page.

Respond in markdown.
"""
```

```python
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
# See how this function creates exactly the format above

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
def summarize(url):
    website = fetch_website_contents(url)

    response = ollama.chat.completions.create(
        model="deepseek-r1:8b",
        messages=messages_for(website)
    )

    return print(response.choices[0].message.content)
```

```python
def display_summary(url):
    summary = summarize(url)
```

```python
print(display_summary("https://quotes.toscrape.com/"))
```
